```python
def mult(a, b):
    n = len(a)
    result = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += a[i][k] * b[k][j]
    return result
```

```python
def np_mult(a, b):
    result = np.dot(a, b)
    return result
```

```python
%%time
M1 = mult(a, b)
```

```python
%%time
M2 = np_mult(a, b)
```

```python
np.abs(np.array(M1) - M2).sum()
```

```python
def scalar_product(a,b):
    s = 0
    s = sum([a[i]*b[i] for i in range(len(a))])
    return s
```

```python
def np_scalar_product(a,b):
    s = 0
    s = np.dot(a, b)
    return s
```

```python
%time product_1 = scalar_product(a,b)
%time product_2 = np_scalar_product(a,b)
np.abs(product_1 - product_2).sum()
```

# **Задание 3**

```python
def np_diag_2k(a):
  arr = np.array(a)
  n = arr.shape[0]
  res = 0

  for i in range(n):
      if arr[i, i] % 2 == 0:
          res += arr[i, i]
  return res
```

```python
a = np.random.randint(1, 10, size=(5, 5))
a
```

```python
%%time
np_diag_2k(a)
```

```python
def np_sec_av(*A):
    result = np.cumsum(A, axis=1)
    return result
```

# **Задание 5**

```python
def np_encode(a):
    elements = []
    counts = []

    current_element = a[0]
    current_count = 1

    for i in range(1, len(a)):
        if a[i] == current_element:
            current_count += 1
        else:
            elements.append(current_element)
            counts.append(current_count)
            current_element = a[i]
            current_count = 1

    elements.append(current_element)
    counts.append(current_count)

    result = np.array(elements), np.array(counts)
    return result
```

```python
from itertools import groupby
X = np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])
%time x, num = np_encode(X)
```

# **Задание 6**

```python
from copy import copy

def np_transformation(X, a):
    """
    param X: array
    """
    x_copy = np.copy(X)

    x_copy[1::2] = a
    x_copy[::2] = np.power(x_copy[::2], 3)

    new_array = np.concatenate((X, x_copy[::-1]))
    result = new_array[::-1]

    return result
```

```python
X = np.array([i for i in range(1, 10, 2)])
%time S2 = np_transformation(X, 5)
```
